# interpolation functions

```python
# univariate interpolation 
from scipy import interpolate
help(interpolate.interp1d)
```

```python
import matplotlib.pyplot as plt
x=np.arange(1,20)
y=np.exp(x/3.0)
f=interpolate.interp1d(x,y)
x1=np.arange(3,12)
y1=f(x1)
plt.plot(x,y,'o',x1,y1,'--')
plt.show()
```
